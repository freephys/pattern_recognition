<h2> Principal Component Analysis </h2>

Suppose we have a data set $D=\{x_1, x_2, ... x_n\}$, where $x_i=(x_{i,1}, x_{i,2}, .. x_{i,d})^t$ is a d-dimnesional vector. We represent data set in matrix form: $[X]_{n\times d}$ with n rows and d columns. The goal of PCA is to find a linear transformation $z = W^t x$ such that new dimensions are completely un-correlated, i.e. the covarainve matrix of transformed data is diagonal. This transformation has the following properties:

* retains most of sample variation
* eigenvectos of the new covariance matrix are orthogonal to each other: $v_i . v_j =0\text{  if }i \neq\ j $
 

We start by calculating the covariance matrix: $$S = cov(X) = \frac{1}{n-1} \sum_{i=1}^{n} (x_k-\bar{x})(x_k-\bar{x})^t = \frac{1}{n-1} X'^t X'$$ where $\bar{x}$ is the sample mean vector, and $X'$ is the centralized form of $X$.

Then, eigenvectors and eigenvalues of $S$ are computed: 
$$eigenvalues = (\lambda_1, \lambda_2, ... \lambda_d )$$
$$eigenvectors = (v_1, v_2, ... v_d)$$

where $\lambda_1 > \lambda_2 > ... > \lambda_d$. Then, we can pick p eigenvectors corresponding to the p highest eigenvalues, and construct the projection matrix $W$:

$$W = \left[ \begin{array}{cccc}
v_{11} & v_{21} & .. & v_{p1} \\
v_{12} & v_{22} & .. & v_{p1} \\
.. & .. & .. & .. \\
v_{1d} & v_{2d} & .. & v_{pd} \\
\end{array} \right]_{d \times p}$$

```python
import numpy as np


d = np.loadtxt("data/2d_linear.txt")
n, dim = d.shape
print(n, dim)

import matplotlib
import matplotlib.pyplot as plt
import pylab
%matplotlib inline

plt.plot(d[:,0], d[:,1], 'go', alpha=0.4)

plt.xlabel('$x_1$', fontsize=18)
plt.ylabel('$x_2$', fontsize=18)
pylab.rc('font', family='serif', size=15)
plt.show()
```

```python
## two ways two calculate the covariance matrix
print(np.cov(d.T))

d_center = np.subtract(d, np.mean(d, axis=0))
S = 1.0/(n-1) * np.dot(d_center.T, d_center)
print(S)
```

```python
## Eigen value decomposition of matrix S:
w, v = np.linalg.eig(S)

print(w)
print(v)
```

Now, using the eigenvector that correponds to the highest eigenvalue, we transform the data into a lower dimensional (1D) space, that has the most variation in data:

```python
print(v.T[0])
d_transformed = np.dot(d_center, v.T)

print(d_transformed.shape)
```

```python
%matplotlib inline

plt.plot(d_transformed[:,0], d_transformed[:,1], 'r^', alpha=0.4)

plt.xlabel('$x_1$', fontsize=18)
plt.ylabel('$x_2$', fontsize=18)
pylab.rc('font', family='serif', size=15)
plt.show()
```
